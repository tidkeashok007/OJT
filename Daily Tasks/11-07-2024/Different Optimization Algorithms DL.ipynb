### Gradient Descent

```python
# Import necessary library
import numpy as np

# Function we want to minimize
def f(x):
    return x**2

# Derivative of the function
def f_prime(x):
    return 2 * x

# Parameters
x = 10  # Starting point
learning_rate = 0.1  # Step size
iterations = 100  # Number of iterations

# Gradient Descent Algorithm
for i in range(iterations):
    gradient = f_prime(x)  # Compute the gradient
    x = x - learning_rate * gradient  # Update x using the gradient

    # Print the value of x and f(x) at each step
    print(f"Iteration {i+1}: x = {x}, f(x) = {f(x)}")
```

### Stochastic Gradient Descent (SGD)

```python
# Import necessary library
import numpy as np

# Function we want to minimize
def f(x):
    return x**2

# Derivative of the function
def f_prime(x):
    return 2 * x

# Parameters
x = 10  # Starting point
learning_rate = 0.1  # Step size
iterations = 100  # Number of iterations

# Stochastic Gradient Descent Algorithm
for i in range(iterations):
    gradient = f_prime(x)  # Compute the gradient
    x = x - learning_rate * gradient  # Update x using the gradient
    
    # Print the value of x and f(x) at each step
    print(f"Iteration {i+1}: x = {x}, f(x) = {f(x)}")
```

### Mini-Batch Gradient Descent

```python
# Import necessary library
import numpy as np

# Function we want to minimize
def f(x):
    return x**2

# Derivative of the function
def f_prime(x):
    return 2 * x

# Parameters
x = 10  # Starting point
learning_rate = 0.1  # Step size
iterations = 100  # Number of iterations
batch_size = 5  # Batch size

# Mini-Batch Gradient Descent Algorithm
for i in range(iterations):
    gradient = np.mean([f_prime(x) for _ in range(batch_size)])  # Compute average gradient over a mini-batch
    x = x - learning_rate * gradient  # Update x using the average gradient

    # Print the value of x and f(x) at each step
    print(f"Iteration {i+1}: x = {x}, f(x) = {f(x)}")
```

### Momentum

```python
# Import necessary library
import numpy as np

# Function we want to minimize
def f(x):
    return x**2

# Derivative of the function
def f_prime(x):
    return 2 * x

# Parameters
x = 10  # Starting point
learning_rate = 0.1  # Step size
iterations = 100  # Number of iterations
momentum = 0.9  # Momentum factor

# Initialize velocity
velocity = 0

# Momentum-based Gradient Descent Algorithm
for i in range(iterations):
    gradient = f_prime(x)  # Compute the gradient
    velocity = momentum * velocity - learning_rate * gradient  # Update velocity
    x = x + velocity  # Update x using the velocity

    # Print the value of x and f(x) at each step
    print(f"Iteration {i+1}: x = {x}, f(x) = {f(x)}")
```

### Adam (Adaptive Moment Estimation)

```python
# Import necessary library
import numpy as np

# Function we want to minimize
def f(x):
    return x**2

# Derivative of the function
def f_prime(x):
    return 2 * x

# Parameters
x = 10  # Starting point
learning_rate = 0.1  # Step size
iterations = 100  # Number of iterations
beta1 = 0.9  # Exponential decay rate for the first moment estimates
beta2 = 0.999  # Exponential decay rate for the second moment estimates
epsilon = 1e-8  # Small number to prevent division by zero

# Initialize first and second moment estimates
m = 0
v = 0

# Adam Optimization Algorithm
for i in range(iterations):
    gradient = f_prime(x)  # Compute the gradient

    # Update biased first moment estimate
    m = beta1 * m + (1 - beta1) * gradient
    # Update biased second moment estimate
    v = beta2 * v + (1 - beta2) * (gradient ** 2)

    # Compute bias-corrected first moment estimate
    m_hat = m / (1 - beta1 ** (i + 1))
    # Compute bias-corrected second moment estimate
    v_hat = v / (1 - beta2 ** (i + 1))

    # Update x using the bias-corrected estimates
    x = x - learning_rate * m_hat / (np.sqrt(v_hat) + epsilon)

    # Print the value of x and f(x) at each step
    print(f"Iteration {i+1}: x = {x}, f(x) = {f(x)}")
```
